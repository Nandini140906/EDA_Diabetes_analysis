since sex is in float value rather than in object we should change it for better understanding
also all the data is in 1,0 form instead of yes or no,lets also convert it for better understanding

```python
def conv(gender):
    if gender==1:
        return "men"
    else:
        return "female"
    
df['Sex']=df['Sex'].apply(conv)
df.head(5)

```

```python
def convert(value):
    if value==1:
        return "yes"
    else:
        return "no"
    


cols=['HighBP','HighChol','Smoker','PhysActivity','Fruits','DiffWalk']  
df[cols]=df[cols].applymap(convert)  #applymap helps to execute every single element as per the instruction being fetched
df.head(10)

def convert1(value):
    if value==0:
        return "no"
    elif value==2:
        return "yes"
    else:
    
        return "no answer"
df['Diabetes_012']=df['Diabetes_012'].apply(convert1)
df
```

VISUALIZATION

```python
diabetic=df[df['Diabetes_012']=='yes']
non_diabetic=df[df['Diabetes_012']=='no'] 
plt.hist(diabetic['BMI'],bins=30,alpha=1.0,label='Diabetic',color='black')
plt.hist(non_diabetic['BMI'],bins=30,alpha=0.6,label='Non-Diabetic',color='pink')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

This shows us that,diabetic individuals generally have higher BMI values compared to non-diabetics, suggesting obesity is a strong risk factor.

VISUALIZATION FOR AGE

```python
diabetic=df[df['Diabetes_012']=='yes']
non_diabetic=df[df['Diabetes_012']=='no']
plt.hist(diabetic['Age'],bins=30,alpha=0.9,color='red',label='Diabetic')
plt.hist(non_diabetic['Age'],bins=30,alpha=0.6,color='pink',label='Non-Diabetic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#Insights from Age vs Diabetes Plot

1.Diabetes is rare in younger groups but rises steadily with age.

2.Middle to older age groups show the highest share of diabetes cases.

3.This confirms age is a strong risk factor, and prevention should focus more on adults as they grow older.

VISUALIZATION FOR GENDER
